```python
X = df.drop('benign_0__mal_1',axis=1).values
y = df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,
    test_size=0.25, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)

#remember to use transform and not fit_transform on test data to avoid
#data leakage
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(30,activation='relu'))
#starting with 30 to match the 30 features of the dataset

model.add(Dense(15,activation='relu'))

#binary classification
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
model.fit(x=X_train,y=y_train,epochs=600,
    validation_data=(X_test,y_test))

#purposefully overfitting to training data
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
#validation and training loss both decrease until around 50, which is 
#good and means that we have yet to overfit.

#However, after this, the training loss continues to decrease while the
#validation loss starts to increase, which indicates that the data is 
#being overfit to the training set. This means we are training to too
#many epochs.
```

```python
model = Sequential()

model.add(Dense(30,activation='relu'))
#starting with 30 to match the 30 features of the dataset

model.add(Dense(15,activation='relu'))

#binary classification
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam')

#Important not to keep training on the same model. You should recreate
#it fresh. We will then use callbacks to stop training before it gets
#out of hand.
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
help(EarlyStopping)
```

```python
#We have to choose a metric to monitor (validation loss), minimum
#change required, patience (number of epochs w/no change before training
#is ended)
```

```python
early_stop = EarlyStopping(monitor='val_loss',mode='min',
                          verbose = 1, patience = 25)

#mode depends on what we are trying to monitor; if we are measuring
#accuracy, we would use max for min, but for validation loss, we want 
#min.
```

```python
model.fit(x=X_train,y=y_train,epochs=600,
    validation_data=(X_test,y_test),callbacks=[early_stop])

#refitting the model and setting a callback parameter equal to the
#early_stop variable, which is always passed as a list 
```

```python
#early stop was triggered and the model stopped training after epoch 
#74 (81 in the lecture)

#we can select an arbitrarily large number of epochs and as long as
#we have callbacks set it doesn't matter
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
#flattening out is ok behavior, but what we want to prevent is the 
#validation loss starting to increase. 
```

```python
#dropout layers turn off a percentage of neurons randomly.
```

```python
#remember to import Dropout from tensorflow.keras.layers if you haven't
```

```python
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()

model.add(Dense(30,activation='relu'))
#starting with 30 to match the 30 features of the dataset
model.add(Dropout(.5))
#rate is probability that you will turn off a neuron; common value is
#between .2 and .5. If you specify a batch size, the dropout percent
#will be the percentage of the batch

model.add(Dense(15,activation='relu'))

#binary classification
model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
model.fit(x=X_train,y=y_train,epochs=600,
    validation_data=(X_test,y_test),callbacks=[early_stop])

#we are now using early stopping and dropouts to prevent overfitting.
```

```python
#it ran for a little bit longer, which means that the dropout layers
#allowed for the model to continue learning longer than it previously 
#had.
```

```python
model_loss = pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

```python
#training and validation loss are continually going down and flatten
#out at essentially the same rate. This is much better than the earlier
#model.
```

```python
predictions = (model.predict(X_test)>.5)*1

#predict_class was removed, so the predict function gives the 
#probability of each class directly. Adding the >.5 argument classifies
#malignancy as a true/false boolean statement, and multiplying by 1
#converts everything to 0s and 1s.
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
#this is very good performance

#accuracy is a good metric since the classes were relatively balanced
```

```python
print (confusion_matrix(y_test,predictions))
```
