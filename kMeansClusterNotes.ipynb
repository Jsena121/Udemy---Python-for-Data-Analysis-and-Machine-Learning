```python
from sklearn.cluster import KMeans
```

```python
#this algorithm starts by randomly assigning each observation to a 
#cluster and finding a centroid to each cluster.
#It then iterates through reassigning datapoints to the cluster whose
#centroid is closest, finds the new centroid, and then repeats this
#over and over until the cluster variation can't be reduced any further
```

```python
kmeans = KMeans(n_clusters = 4)

#we know there are 4 clusters beforehand bc we made the data, but what
#about when we don't know?
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1])

#plot shows the centers of each cluster
```

```python
kmeans.labels_

#shows the labels it believes to be true for the clusters
```

```python
#if you did not know the labels you would be done, but we do know the 
#labels in data[1]
```

```python
data[1]
```

```python
#kmeans is unsupervised so it is not used to predict labels but rather
#to find them
```

```python
fig , (ax1,ax2) = plt.subplots(1,2, sharey=True,figsize = (10,6))

#creates a subplot of one row by two columns, sharey=True allows them
#to share the axes. Tuple unpacking is used (ax1,ax2) to grab these
#two sets of axes

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')
#plots the original data and colors it based on what the algorithm
#thinks the labels should be

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
#plots the original data and colors them by their actual blobs.
```

```python
#there was some noise in the original group where the clusters are 
#close together that kmeans couldn't deal with and assigned the data
#points to the same central cluster (note the overlap in the orignal
#between red,tan, and purple that is not present in kmeans)

#difference between how each is plotted can also be disregarded (yellow
#and purple are swapped)
```
